```python
#importamos la librerias
import tensorflow as tf
import numpy as np
```

```python
#definir las entradas y salidas
celsius=np.array([-40,-10,0,8,15,22,38], dtype=float)
fahrenheit=np.array([-40,14,32,46,59,72,100],dtype=float)
```

```python
#crear modelo neuronal
capa=tf.keras.layers.Dense(units=1, input_shape=[1])
modelo=tf.keras.Sequential([capa])
```

```python
#definiendo el modelo a compilar
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
#entrenamiento
print("Comenzado el entrenamiento")
historial=modelo.fit(celsius, fahrenheit,epochs=1000,verbose=False)
print("Termino el entrenamiento")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoc")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])
```

```python
#Comprobar si funciona el modelo
print("Hagamos una prediccion")
resultado=modelo.predict([100])
print("El resultado en Fahrenheit es " + str(resultado))
```
